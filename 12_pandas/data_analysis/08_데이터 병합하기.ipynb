### 데이터 병합하기

```python
import pandas as pd

df=pd.read_csv("./data/seoul_park04.csv")
df.head()
```

```python
df2=pd.read_csv("./data/seoul_park_april.csv")
df2.head()
```

### 데이터 병합: concat()

```python
# axis : 0은 행 방향 결합, 1은 열 방향 결합(기본값은 0)
# join='inner' -> 공통 컬럼만 유지(기본값은 outer)
# ignore_index=True -> 기존 인덱스 무시하고 새로 인덱스를 만듦(기본값은 False)
pd.concat([df, df2], axis=0, join='inner', ignore_index=True)
```

### 데이터 병합: merge()

```python
mm=pd.read_csv("./data/misemunji.csv")
mm.head()
```

```python
df["날짜"]=pd.to_datetime(df["날짜"])
mm["날짜"] = pd.to_datetime(mm["날짜"], format="%Y%m%d")
mm["날짜"]
```

```python
# on : 데이터프레임을 합질 때 기준이 되는 컬럼의 이름
# how : 데이터 프레임의 데이터를 합치는 방법(inner, outer, left, right)
# inner : 교집합(두 데이터프레임이 모두 존재하는 데이터만)
# outer : 합집합(두 데이터프레임의 모든 데이터 사용)
# left : 왼쪽 데이터프레임에 있는 모든 데이터
# right : 오른쪽 데이터프레임에 있는 모든 데이터 
pd.merge(df, mm, on='날짜', how='left')
```
