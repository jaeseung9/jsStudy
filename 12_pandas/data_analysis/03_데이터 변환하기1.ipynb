### 데이터 변환하기1

```python
import pandas as pd

# 파일 불러오기
df=pd.read_csv('./data/seoul_park.csv')
df.head()
```

```python
df
```

```python
df.info()
```

```python
# 텍스트 데이터를 정수형으로 바꾸기 전에 텍스트에서 숫자만 남기고 제거한다.
# 이를 위해 숫자로 바꾸어야 하는 컬럼들에서 replace()를 활용하여 -를 0으로 바꾸고 콤마를 제거한다.

# str.replace()
columns=['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in columns:
    df[i]=df[i].str.replace('-', '0')
    df[i]=df[i].str.replace(',', '')
```

### 데이터 타입 변환: to_numeric()

```python
# 숫자로만 변환 가능
df["유료합계"]=pd.to_numeric(df["유료합계"])

df.info()
```

```python
## df의 "외국인" 컬럼의 자료형을 숫자형으로 변환해 변수 foreigner에 저장하세요
foreigner = df["외국인"].astype(int)
foreigner.info()
```

```python
columns=['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in columns:
    df[i]=pd.to_numeric(df[i])

df.info()
```

### 데이터 타입 변환: to_datetime()

```python
df["날짜"]=pd.to_datetime(df["날짜"])
df.info()
```

```python
# datetime 형태의 데이터에는 연, 월, 일 정보가 담겨있어 이를 활용하여 새로운 정보를 생성할 수 있다.
df['연']=df['날짜'].dt.year
df['월']=df['날짜'].dt.month
df['일']=df['날짜'].dt.day
df['요일']=df['날짜'].dt.dayofweek  # 요일은 숫자로 저장(월~일:0~6)
```
