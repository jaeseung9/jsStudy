### 데이터 추출하기

```python
import pandas as pd

df = pd.read_csv("./data/seoul_park03.csv")
df.head()
```

```python
### 조건에 따른 인덱싱: Boolean indexing
df[df["어른"] > 5000]
```

```python
# pandas 논리연산자 사용
# 두 개 이상의 조건은 and(&)나 or(|)을 활용하여  같이 사용한다. 이 때 각 조건식은 소괄호로 묶여있어야 한다.

# 어른 입장객 수가 10000명이 넘는 공휴일의 데이터만을 추출
df[(df["어른"] > 10000) & (df["요일"] == "공휴일")]
```

```python
# 어른 입장객 수가 10000명이 넘거나 어린이 입장객 수가 2000명이 넘는 날짜의 데이터 추출
df[(df["어른"] > 10000) | (df["어린이"] > 2000)]
```

```python
# 어린이날의 데이터를 추출해 childrensday에 저장하세요
# "월"과 "일" 컬럼을 사용하세요.abs

childrensday = df[(df["월"] == 5 ) & (df["일"] == 5)]
childrensday
```

```python
# 2026년 1월 4일의 어른 데이터 추출
# .loc[행의 인덱스, 열의 인덱스]
df.loc[3,"어른"]
```

```python
# 슬라이싱을 활용한 인덱싱
# 유의할점 : loc는 라벨 기반 인덱싱이기 때문에 A:B로 슬라이싱을 하면 A부터 B까지, 즉 B를 포함한 범위를 인덱싱 한다.abs

df.loc[3:6,"어른":"외국인"]
```

```python
### loc와 Boonlean indexing을 활용한 데이터 추출
```

```python
# 어른과 어린이의 입장객 수가 둘 다 1000보다 큰 날짜의 "날짜"부터 "총계" 행 추출
df.loc[(df["어른"]> 1000) & (df["어린이"] > 1000), "날짜" : "총계"]
```
