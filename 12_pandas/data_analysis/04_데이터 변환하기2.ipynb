### 데이터 변환하기2

```python
import pandas as pd

df = pd.read_csv("./data/seoul_park02.csv")
df.head()
```

### 열 전체 변환: map()

```python
week={0:'월',1:'화',2:'수',3:'목',4:'금',5:'토',6:'일'}
```

```python
df['요일'] = df['요일'].map(week)
df.head()
```

```python
df['날씨'].value_counts()
```

```python
def weather(e):
    if e == '눈' or e == '비':
        return '눈/비'
    else:
        return e
```

```python
df['날씨'] = df['날씨'].apply(weather)
```

```python
df['날씨'].value_counts()
```

```python
# 람다 함수 적용
df["날씨"] = df["날씨"].apply(lambda e : "눈/비" if e == "눈" or e == "비" else e)
```

```python
# 일차별 혼잡도 시리즈를 생성한 뒤 변수 congestion에 저장하세요.

# apply()를 사용하세요
# '총계'컬럼의 값이 8000 이상인 날은 "혼잡", 8000미만인 날은 "원활"로 합니다.
# 평범하게 함수를 정의하거나, 람다함수를 활용해 보세요.

def check(x):
    if x >= 8000:
        return "혼잡"
    else x
        return "원활"
        

congestion = df["총계"].apply(check)
```
