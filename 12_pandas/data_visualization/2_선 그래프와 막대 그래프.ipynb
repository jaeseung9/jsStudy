### 선 그래프와 막대 그래프

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="Malgun Gothic")
```

```python
import pandas as pd
```

```python
df=pd.read_csv("seoul_park.csv")
df.head()
```

```python
df.info()
```

### 1. 선 그래프

#### 1-1. Matplotlib

월별 데이터 묶기(groupby() 사용)

```python
gb_mon = df.groupby("월")
```

```python
df_mon = gb_mon.mean(numeric_only=True)  # 숫자형 열만 평균을 계산한다.
df_mon
```

```python
# 어른, 어린이, 청소년별 월별 입장객수 선그래프로 보기
plt.title("서울대공원 입장객 분석", fontsize=20)    # 그래프 제목
plt.plot(df_mon["어른"], label="어른")
plt.plot(df_mon["어린이"], label="어린이")
plt.plot(df_mon["청소년"], label="청소년")
plt.xlabel("월", fontsize=15)
plt.ylabel("입장객수", fontsize=15)
plt.legend()
plt.show()
```

```python
# 어른, 어린이, 청소년별 월별 입장객수 선그래프로 보기
plt.title("서울대공원 입장객 분석", fontsize=20)    # 그래프 제목
plt.plot(df_mon["어른"], label="어른")
plt.plot(df_mon["어린이"], label="어린이")
plt.plot(df_mon["청소년"], label="청소년")
plt.xlabel("월", fontsize=15)
plt.ylabel("입장객수", fontsize=15)
plt.legend()
plt.xticks([mon for mon in range(1, 13)])   # x축 눈금 조절 : 1부터 12가 들어있는 리스트를 xticks에 전달
plt.show()
```

### 선 그래프 모양 변경하기

1. 마커(marker) : 그래프의 데이터 포인트를 나타내는 작은 심볼
2. 선 모양(linestyle) : 선의 모양을 설정한다.
3. 색(color) : 그래프의 색을 설정

```python
plt.title("서울대공원 입장객 분석", fontsize=20)    # 그래프 제목
plt.plot(df_mon["어른"], marker=".", linestyle="-", color="b", label="어른")
plt.plot(df_mon["어린이"], marker="o", linestyle="--", color="k", label="어린이")
plt.plot(df_mon["청소년"], marker="^", linestyle=":", color="r", label="청소년")
plt.xlabel("월", fontsize=15)
plt.ylabel("입장객수", fontsize=15)
plt.grid(True, axis="x")      # x 축에만 격자 생성
plt.legend()
plt.xticks([mon for mon in range(1, 13)])   # x축 눈금 조절 : 1부터 12가 들어있는 리스트를 xticks에 전달
plt.show()
```

### seaborn

data 매개변수에 시각화하려는 데이터 프레임 전달하고 
x축과 y축에 들어갈 칼럼의 이름을 x, y 매개변수에 전달하여 간단하게 그래프를 그릴 수 있다.

```python
sns.lineplot(data=df, x="월", y="어른", label="어른", errorbar=None) # 기본 통계는 mean이며, estimator 옵션을 통해 바꿀 수 있다.
# 시본은 통계 시각화 라이브러리이다. 집계 + 시각화를 한번에 처리함
# 위 코드에서 시본은 내부적으로 하는 일은
# df.groupby("월")["어른"].mean()
sns.lineplot(data=df, x="월", y="어린이", label="어린이", errorbar=None)
sns.lineplot(data=df, x="월", y="청소년", label="청소년", errorbar=None)
plt.ylabel("입장객수")
plt.xticks([mon for mon in range(1, 13)])

plt.show()
```

### 막대 그래프

#### Matplotlib

```python
df_week = df.groupby("요일").mean(numeric_only=True)
df_week = df_week.reset_index() # '요일'열을 다시 데이터프레임의 열로 가져온다.
df_week
```

```python
# 막대 그래프
plt.title("서울대공원 청소년 입장객 분석" ,fontsize=20)
plt.bar(df_week["요일"], df_week["청소년"])
plt.xlabel("요일", fontsize=15)
plt.ylabel("청소년 입장객수", fontsize=15)
```

```python
# 가로 막대 그래프
plt.title("서울대공원 청소년 입장객 분석" ,fontsize=20)
plt.barh(df_week["요일"], df_week["청소년"])
plt.xlabel("청소년 입장객수", fontsize=15)
plt.ylabel("요일", fontsize=15)
```

```python
# 데이터 프레임 정렬 후 시각화
df_sorted = df_week.sort_values(by="청소년", ascending=False)

plt.title("서울대공원 청소년 입장객 분석")
plt.bar(df_sorted["요일"], df_sorted["청소년"])
plt.xlabel("요일")
plt.ylabel("청소년 입장객수")
```

#### Seaborn

```python
sns.barplot(data=df, x="요일", y="청소년", errorbar=None)
```

```python
# 날씨별 어린이 입장객의 수를 막대그래프로
sns.barplot(data=df, x="날씨", y="어린이", errorbar=None)
```

```python
sns.barplot(data=df, x="날씨", y="어린이", errorbar=None, hue="공휴일") # "공휴일" 값을 기준으로 그룹을 나눔
```

### 연도별 매출액 그래프 그리기

다음 조건을 모두 만족하도록 그래프를 그려보세요.

- `plt.subplots`을 이용해 1행 2열로 구성된 서브플롯을 생성하여 `ax1`, `ax2`에 저장합니다.
- `ax1`에는 x축을 `"연"`, y축을 `"매출액"`으로 하는 선 그래프를 그립니다. (`sns.lineplot` 사용)
    - `errorbar=None`을 전달합니다.
    - 마커는 `"o"`로 지정합니다.
    - 색은 `"r"`로 지정합니다.
    - `ax1`의 제목은 `"연도별 매출액"`으로 설정합니다.
- `ax2`에는 x축을 `"연"`, y축을 `"매출액"`으로 하는 막대 그래프를 그립니다. (`sns.barplot` 사용)
    - `errorbar=None`을 전달합니다.
    - `hue`값을 이용하여 `"공휴일"`을 기준으로 그룹을 나눠 그립니다.
    - `ax2`의 제목은 `"공휴일 여부에 따른 매출액"`으로 설정합니다.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)

sns.lineplot(data=df, x="연", y="매출액", errorbar=None, marker="o", color="r", ax=ax1)
ax1.set_title("연도별 매출액")

sns.barplot(data=df, x="연", y="매출액", errorbar=None, hue="공휴일", ax=ax2)
ax2.set_title("공휴일 여부에 따른 매출액")

plt.show()
```
