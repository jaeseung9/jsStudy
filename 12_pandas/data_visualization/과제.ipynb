```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "San Francisco"],
    "Salary": [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
df[["Name", "City"]]
```

```python
df[df["Age"] >= 30]
```

```python
df["Country"] = "USA"
df
```

```python
df.loc[len(df)] = ["Frank", 33, "Seattle", 85000, "USA"]
df
```

```python
df = df.drop(columns=["Salary"])
df
```

```python
df = df[df["Name"] != "David"]
df
```

```python
df["Salary"] = 5000

df.loc[df["Age"] >= 35, "Salary"] *= 1.05

df
```

```python
df.loc[df["Name"] == "Alice", "City"] = "Boston"
df
```

```python
df["Age"] = df["Age"].astype(float)
df["Salary"] = df["Salary"].astype(float)

df.dtypes
```

```python
employee_data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['John', 'Emily', 'Sara', 'Tom', 'Anna'],
    'Department': ['HR', 'IT', 'Marketing', None, 'HR'],
    'Salary': [50000, 70000, None, 45000, 52000],
    'YearsAtCompany': [5, None, 8, 3, 4]
}

df_emp = pd.DataFrame(employee_data)

df_emp["Salary"] = df_emp["Salary"].fillna(df_emp["Salary"].mean())
df_emp["YearsAtCompany"] = df_emp["YearsAtCompany"].fillna(df_emp["YearsAtCompany"].mean())
df_emp
```

```python
product_data = {
    'ProductID': [1, 2, 3, 1, 4, 2],
    'ProductName': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Keyboard', 'Mouse'],
    'Price': [1200, 25, 300, 1150, 45, 22],
    'QuantitySold': [10, 100, 50, 8, 70, 120]
}

df_prod = pd.DataFrame(product_data)

df_prod_sorted = df_prod.sort_values(by="Price", ascending=True)

df_prod_clean = df_prod_sorted.drop_duplicates(
    subset=["ProductID"],
    keep="first"
)

df_prod_clean
```

```python
car_data = {
    'Model': ['Sedan', 'SUV', 'Truck', 'Coupe', 'Convertible'],
    'Horsepower': [180, 250, 200, 170, 150],
    'MPG': [30, 20, 15, 35, 28],
    'Price': [30000, 45000, 40000, 28000, 32000]
}

df_car = pd.DataFrame(car_data)

low_hp3 = df_car.sort_values(by="Horsepower", ascending=True).head(3)

result = low_hp3.sort_values(by="MPG", ascending=False)

result

```
