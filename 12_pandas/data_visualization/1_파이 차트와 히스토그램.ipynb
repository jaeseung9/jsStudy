### 파이 차트
전체 데이터에서 그 범주에 해당하는 비율을 표현한다. 
matplotlib의 pie함수를 사용(seaborn에는 별도로 없음)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="Malgun Gothic")
```

```python
df = pd.read_csv("seoul_park.csv")
df.head()
```

```python
# 2019년의 어른, 청소년, 어린이 입장 분포 보기
data_2019 = df[df["연"]==2019][["어른", "어린이", "청소년"]].sum()
data_2019
```

```python
# 파이차트에는 데이터와 레이블이 필요하다.
# 데이터는 각 범주의 숫자를 리스트로 전달하면 된다.
# 레이블은 각 비율이 어떤 항목의 비율인지 보여주기 위해 필요하다.

labels = data_2019.index
data = data_2019
data
```

```python
# 파이차트 그리기
plt.pie(x=data, labels=labels, autopct="%1.1f%%")   # 소수점 첫째자리까지 표시
plt.legend(loc="lower left")
plt.show()
```

### 히스토그램
특정 구간에 해당하는 빈도를 표현하는 그래프

#### Matplotlib

plt.hist()

```python
# 토요일 어린이 입장객 수의 분포를 히스토그램으로 시각화
df_sat = df[df["요일"]=="토"]
df_sat

plt.hist(df_sat["어린이"], bins=15, histtype="stepfilled") # bins : 몇개의 구간으로 나눌 것인가 -> 막대의 수
plt.show()
df_sat
```

### seborn
sns.histplot()

```python
sns.histplot(data=df_sat, x="어른")
```

```python
# 연도별 2월 어린이 입장객 수 비교
df_feb = df[df["월"]==2]
sns.histplot(data=df_feb, x="어린이", hue="연", bins = 20, binrange=(0, 600)) # binrange=(최소, 최대) -> 구간의 범위 설정
```

```python
# multiple 설정
sns.histplot(
    data=df_feb, 
    x="어린이", 
    hue="연", 
    bins = 10, 
    binrange=(0, 800),
    multiple="dodge",    # 그래프가 겹치지 않도록 설정
    shrink=0.8  # 막대의 너비를 조절
)
```
