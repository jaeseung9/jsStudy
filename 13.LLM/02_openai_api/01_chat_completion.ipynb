# Chat Completion

```python
# 환경설정

from dotenv import load_dotenv

load_dotenv()
```

```python
import os

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "당신은 친절하고, 상세한 설명을 잘하는 챗봇입니다."},
        {"role": "user", "content": "안녕하세요, 제 이름은 유재석입니다."},
        {"role" : "assistant", "content" : "안녕, 재석아! 만나서 반가워! 어떻게 지내고 있어 궁금한 점이나 이야기하고 싶은 것이 있으면 언제든지 말해줘!"},
        {"role": "user", "content" : "내 이름이 뭐라고??"}
    ],
    temperature=1,
    max_tokens=4096,
    top_p=1
)

print(response.choices[0].message.content)
```
