### TTS(Text-to-Speech)
텍스트를 자연스러운 음성으로 변환하는 AI 모델이다.

** 음성 선택지 **
- ash
- ballad
- sage
- marin
- cedar
- fable
- coral

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI

client = OpenAI()

input = "화요일인 23일 오후부터 24일 새벽 사이 전국 대부분 지역에 비가 내리겠다. 경기북부와 강원내륙·산지에는 비 또는 눈이 올 수 있다. 일부 경상권은 24일 아침까지, 제주도는 오후까지 비가 이어지는 곳이 있겠다."

with client.audio.speech.with_streaming_response.create(
    model="tts-1",
    voice="nova",
    input=input[:200]
) as response:
    response.stream_to_file("output.mp3")
    print("파일 생성이 완료되었습니다.")
```
