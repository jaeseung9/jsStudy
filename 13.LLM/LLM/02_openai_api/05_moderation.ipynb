```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI
import pandas as pd

client = OpenAI()

response = client.moderations.create(
    model="omni-moderation-latest",
    input="저놈은 개새끼입니다."
)

moderation_df = pd.DataFrame(response.results[0].categories, columns=['category', 'bool'])
moderation_df
```
